```python
#Remove clusters below a threshold on the number of replicates
lU,counts = np.unique(labels,return_counts=True)
lU = lU[counts>3]
idx = np.hstack([np.where(labels==l)[0] for l in lU])

matrix = matrix[idx,:]
labels = labels[idx]
names = names[idx]
namesCT = namesCT[np.unique(labels)]

labels_tmp = labels.copy()
for i,l in enumerate(np.unique(labels)):
    labels_tmp[labels==l] = i
labels = labels_tmp
```

```python
#Reorder the labels by sorting by names
idx = np.argsort(namesCT)
labels_tmp = labels.copy()
namesCT_tmp = namesCT.copy()

for i in range(labels.max()+1):
    labels_tmp[labels==i]=idx[i]
    namesCT_tmp[i] = namesCT[idx[i]]

labels = labels_tmp
namesCT = namesCT_tmp
```

```python
#Reorder the labels
labels_tmp = labels.copy()
namesCT_tmp = namesCT.copy()

lU,idx = np.unique(labels,return_index=True)
idxU = np.argsort(idx)

for i,l in enumerate(lU[idxU]):
    labels_tmp[labels==l]=i
    namesCT_tmp[l] = namesCT[i]

labels = labels_tmp
namesCT = namesCT_tmp
```
